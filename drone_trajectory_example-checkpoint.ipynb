```python

# ----------------------------
# 0) Global settings & timeline
# ----------------------------
dt = 0.1
T  = 2.2
tsim = np.arange(0, T + dt/2, step=dt)
N = len(tsim)

# ----------------------------
# 1) Environment: wind & altitude
# ----------------------------
w_mag = np.round(np.random.uniform(1.0, 7.5), 2)            # wind speed [m/s]
w = w_mag * np.ones_like(tsim)
zeta_deg = np.round(np.random.uniform(0.0, 360.0), 2)       # wind dir [deg]
zeta = np.deg2rad(zeta_deg) * np.ones_like(tsim)            # rad
z = 2.0 * np.ones_like(tsim)                                # altitude (const)

# ----------------------------
# 2) Speed profile g(t)
# ----------------------------
min_speed, max_speed = 1.6, 7.5  # m/s
speed_behavior = np.random.choice(['sinusoidal', 'linear', 'multiple_linear', 'rapid_jagged'])
initial_speed = np.round(np.random.uniform(min_speed, max_speed), 2)

if speed_behavior == 'sinusoidal':
    low_speed  = np.random.uniform(min_speed, (min_speed + max_speed) / 2)
    high_speed = np.random.uniform((min_speed + max_speed) / 2, max_speed)
    midpoint_speed = 0.5 * (low_speed + high_speed)
    amplitude      = 0.5 * (high_speed - low_speed)
    frequency_hz   = np.random.uniform(0.1, 1.2)  # Hz
    phase_offset   = np.random.uniform(0, 2*np.pi)
    g = midpoint_speed + amplitude * np.sin(2*np.pi*frequency_hz * tsim + phase_offset)

elif speed_behavior == 'linear':
    g = np.full_like(tsim, initial_speed, dtype=float)
    target_speed   = np.round(np.random.uniform(min_speed, max_speed), 2)
    ramp_duration  = np.random.uniform(0.3*T, 0.9*T)
    ramp_start_time = np.random.uniform(0.0, T - ramp_duration)
    ramp_end_time   = ramp_start_time + ramp_duration
    ramp_mask = (tsim >= ramp_start_time) & (tsim <= ramp_end_time)
    if ramp_mask.any():
        g[ramp_mask]  = np.linspace(initial_speed, target_speed, ramp_mask.sum())
        g[tsim > ramp_end_time] = target_speed

elif speed_behavior == 'multiple_linear':
    num_segments_spd  = np.random.randint(2, 6)  # 2..5 segments
    segment_times_spd = np.sort(np.r_[0.0, np.random.uniform(0, T, size=num_segments_spd-1), T])
    segment_speeds    = np.round(np.random.uniform(min_speed, max_speed, size=segment_times_spd.size), 2)
    g = np.interp(tsim, segment_times_spd, segment_speeds)

elif speed_behavior == 'rapid_jagged':
    num_segments_spd   = np.random.randint(2, 7)  # 2..6 segments
    boundaries_spd     = np.sort(np.r_[0.0, np.random.uniform(0, T, size=num_segments_spd-1), T])
    g = np.empty_like(tsim, dtype=float)
    for i in range(len(boundaries_spd) - 1):
        segment_speed = np.round(np.random.uniform(min_speed, max_speed), 2)
        if i < len(boundaries_spd) - 2:
            mask = (tsim >= boundaries_spd[i]) & (tsim < boundaries_spd[i+1])
        else:
            mask = (tsim >= boundaries_spd[i]) & (tsim <= boundaries_spd[i+1])
        g[mask] = segment_speed

g = np.clip(g, min_speed, max_speed)

# ==============================================================
# 3) Path mode: HYBRID vs SOLO (SOLO = one primitive for whole path)
#    'solo' occurs 15%; hybrid otherwise. PureAngles is now a primitive
#    option (not a separate mode). Hybridization count +1 (more switches).
# ==============================================================
path_mode = np.random.choice(['hybrid', 'solo'], p=[0.85, 0.15])

theta = np.zeros(N, dtype=float)
primitive_names = []

# --- helpers for segment partitioning ---
min_seg_time = 0.3
min_seg_len  = max(2, int(np.round(min_seg_time / dt)))

# heavier bias buckets for direction changes (deg)
angle_buckets = [(10, 40), (45, 100), (110, 200), (220, 360)]
bucket_weights = np.array([0.36, 0.32, 0.22, 0.10])
bucket_probs = bucket_weights / bucket_weights.sum()

# --- full primitive catalog (now includes PureAngles and all new ones) ---
primitive_types  = np.array([
    # classics
    'Straight', 'DirectTurn', 'HybridStraightTurn', 'Arc', 'Clothoid',
    'Loop', 'Spiral', 'SnailSpiral', 'ZigzagDrift', 'UTurn', 'BrokenArcChain',
    'PureAngles',
    # new:
    'RandomPolygon', 'DubinsRS', 'BezierSpline', 'FigureEight', 'RoseCurve',
    'CycloidFamily', 'OUCurvature', 'CorridorWeave', 'PursuitEvasion'
])

primitive_weights = np.array([
    0.9,  # Straight
    1.1,  # DirectTurn
    2.0,  # HybridStraightTurn
    1.1,  # Arc
    1.0,  # Clothoid
    1.5,  # Loop
    1.0,  # Spiral
    1.8,  # SnailSpiral
    1.0,  # ZigzagDrift
    1.2,  # UTurn
    1.0,  # BrokenArcChain
    1.4,  # PureAngles
    # new ones:
    1.4,  # RandomPolygon
    1.3,  # DubinsRS
    1.2,  # BezierSpline
    1.3,  # FigureEight
    1.2,  # RoseCurve
    1.3,  # CycloidFamily
    1.4,  # OUCurvature
    1.2,  # CorridorWeave
    1.1   # PursuitEvasion
])
primitive_probs = primitive_weights / primitive_weights.sum()

# --- choose number of HYBRID segments (more than before) ---
if path_mode == 'hybrid':
    # previously ~1..5; now bump by +1 => 2..6
    num_segments_path = np.random.randint(2, 7)
else:
    num_segments_path = 1

base = np.full(num_segments_path, min_seg_len, dtype=int)
remaining = N - base.sum()
if remaining < 0:
    base = np.ones(num_segments_path, dtype=int); remaining = N - base.sum()
if remaining > 0:
    weights = np.random.rand(num_segments_path); weights /= weights.sum()
    extra = (weights * remaining).astype(int)
    while extra.sum() < remaining:
        extra[np.random.randint(0, num_segments_path)] += 1
    seg_lens = base + extra
else:
    seg_lens = base.copy()
seg_lens[-1] += (N - seg_lens.sum())

seg_starts = np.cumsum(np.r_[0, seg_lens[:-1]])
seg_ends   = seg_starts + seg_lens

# --- build theta over segments ---
angle_current = np.random.uniform(-np.pi, np.pi)
angle_turn_list = []

for k in range(num_segments_path):
    Lk = int(seg_lens[k])
    tloc = np.arange(Lk) * dt
    ptype = np.random.choice(primitive_types, p=primitive_probs) if path_mode == 'hybrid' else np.random.choice(primitive_types, p=primitive_probs)
    if path_mode == 'solo':
        # keep the same ptype for all time
        if k == 0:
            ptype_solo = ptype
        else:
            ptype = ptype_solo

    # draw a bucketed change amplitude for primitives that need a net change
    bucket = angle_buckets[np.random.choice(len(angle_buckets), p=bucket_probs)]
    ddeg   = np.random.uniform(bucket[0], bucket[1])
    sign   = 1.0 if np.random.rand() < 0.5 else -1.0
    dtheta = np.deg2rad(ddeg) * sign

    # when needed, encourage tightness
    tight_boost = 1.0
    if ptype in ['Arc', 'Clothoid', 'BrokenArcChain', 'Loop', 'FigureEight', 'RoseCurve', 'CycloidFamily'] and np.random.rand() < 0.35:
        tight_boost = np.random.uniform(1.1, 1.6)

    # ----------------- primitive implementations (theta_seg) -----------------
    if ptype == 'Straight':
        # keep but avoid too-long monotone by injecting slight drift
        drift = np.deg2rad(np.random.uniform(-8, 8))
        theta_seg = np.full(Lk, angle_current + drift)

    elif ptype == 'DirectTurn':
        theta_seg = np.full(Lk, angle_current + dtheta)
        angle_current = theta_seg[-1]

    elif ptype == 'HybridStraightTurn':
        min_split = max(1, int(0.2 * Lk))
        split_idx = np.random.randint(min_split, Lk - min_split + 1) if Lk > 2 else 1
        corner_deg = np.random.uniform(5, 175)
        corner_sign = 1.0 if np.random.rand() < 0.5 else -1.0
        corner_dtheta = np.deg2rad(corner_deg) * corner_sign
        theta_seg = np.empty(Lk, dtype=float)
        theta_seg[:split_idx] = angle_current
        angle_current = angle_current + corner_dtheta
        blend = min(3, max(1, int(0.05 * Lk)))
        if split_idx + blend < Lk:
            blend_t = np.linspace(0, 1, blend, endpoint=False)
            theta_seg[split_idx:split_idx+blend] = angle_current * blend_t + theta_seg[split_idx-1] * (1 - blend_t)
            theta_seg[split_idx+blend:] = angle_current
        else:
            theta_seg[split_idx:] = angle_current
        angle_turn_list.append(np.round(corner_deg * corner_sign, 1))

    elif ptype == 'Arc':
        if Lk > 1:
            theta_seg = angle_current + (dtheta * tight_boost) * (tloc / tloc[-1])
        else:
            theta_seg = np.array([angle_current + dtheta * tight_boost])
        angle_current = theta_seg[-1]

    elif ptype == 'Clothoid':
        if Lk > 1:
            s = 0.5 * (1 - np.cos(np.pi * tloc / tloc[-1]))
        else:
            s = np.array([1.0])
        theta_seg = angle_current + (dtheta * tight_boost) * s
        angle_current = theta_seg[-1]

    elif ptype == 'Loop':
        n_loops = np.random.choice([1,2,3,4], p=np.array([0.25,0.4,0.25,0.10]))
        total = 2*np.pi * n_loops * (1.0 if np.random.rand() < 0.5 else -1.0)
        if Lk > 1:
            theta_seg = angle_current + (total * tight_boost) * (tloc / tloc[-1])
        else:
            theta_seg = np.array([angle_current + total * tight_boost])
        angle_current = theta_seg[-1]

    elif ptype == 'Spiral':
        revs = np.random.uniform(0.5, 2.5) * (1.0 if np.random.rand() < 0.5 else -1.0)
        total = 2*np.pi * revs
        if Lk > 1:
            tend = tloc[-1]
            w0 = np.random.uniform(-2.0*np.pi, 2.0*np.pi)
            a  = 2*(total - w0*tend) / (tend**2 + 1e-9)
            theta_seg = angle_current + w0*tloc + 0.5*a*(tloc**2)
        else:
            theta_seg = np.array([angle_current + total])
        angle_current = theta_seg[-1]

    elif ptype == 'SnailSpiral':
        revs = np.random.uniform(1.5, 4.0) * (1.0 if np.random.rand() < 0.5 else -1.0)
        total = 2*np.pi * revs
        if Lk > 1:
            tend = tloc[-1]
            a_sign = 1.0 if np.random.rand() < 0.65 else -1.0
            a_mag  = np.random.uniform(0.05, 0.35) * np.pi
            a = a_sign * a_mag
            w0 = (total - 0.5*a*(tend**2)) / (tend + 1e-9)
            theta_seg = angle_current + w0*tloc + 0.5*a*(tloc**2)
        else:
            theta_seg = np.array([angle_current + total])
        angle_current = theta_seg[-1]

    elif ptype == 'ZigzagDrift':
        drift_rate = np.random.uniform(-0.6, 0.6)
        amp        = np.deg2rad(np.random.uniform(25, 85))
        freq_hz    = np.random.uniform(0.25, 1.2)
        phase0     = np.random.uniform(-np.pi, np.pi)
        theta_seg  = angle_current + drift_rate*tloc + amp*np.sin(2*np.pi*freq_hz*tloc + phase0)
        angle_current = theta_seg[-1]

    elif ptype == 'UTurn':
        dtheta_u = (np.pi * (1.0 if np.random.rand() < 0.5 else -1.0)) * np.random.uniform(0.8, 1.2)
        if Lk > 1:
            s = 0.5 * (1 - np.cos(np.pi * tloc / tloc[-1]))
        else:
            s = np.array([1.0])
        theta_seg = angle_current + dtheta_u * s
        angle_current = theta_seg[-1]

    elif ptype == 'BrokenArcChain':
        n_sub = np.random.randint(2, 4)
        base_sub = np.full(n_sub, max(1, Lk // n_sub), dtype=int)
        base_sub[-1] += (Lk - base_sub.sum())
        theta_seg = np.empty(0, dtype=float)
        angle_here = angle_current
        cur_sign = 1.0 if np.random.rand() < 0.5 else -1.0
        for j in range(n_sub):
            Lj = int(base_sub[j]); tj = np.arange(Lj) * dt
            bkt = angle_buckets[np.random.choice(len(angle_buckets), p=bucket_probs)]
            ddeg_j = np.random.uniform(bkt[0], bkt[1])
            dth_j  = np.deg2rad(ddeg_j) * cur_sign * tight_boost
            if Lj > 1:
                seg_j = angle_here + dth_j * (tj / tj[-1])
            else:
                seg_j = np.array([angle_here + dth_j])
            angle_here = seg_j[-1]
            theta_seg = np.concatenate([theta_seg, seg_j])
            cur_sign *= -1.0
        angle_current = theta_seg[-1]

    elif ptype == 'PureAngles':
        # 2–4 straight legs inside this segment, sharp corners (5–175°)
        n_legs = np.random.randint(2, 5)
        base_sub = np.full(n_legs, max(1, Lk // n_legs), dtype=int); base_sub[-1] += (Lk - base_sub.sum())
        theta_seg = np.empty(0, dtype=float)
        angle_here = angle_current
        for j in range(n_legs):
            Lj = int(base_sub[j])
            seg = np.full(Lj, angle_here)
            theta_seg = np.concatenate([theta_seg, seg])
            if j < n_legs - 1:
                corner_deg = np.random.uniform(5, 175)
                corner_sign = 1.0 if np.random.rand() < 0.5 else -1.0
                angle_here = angle_here + np.deg2rad(corner_deg) * corner_sign
                angle_turn_list.append(np.round(corner_deg * corner_sign, 1))
        angle_current = theta_seg[-1]

    elif ptype == 'RandomPolygon':
        # 3–8 straight edges, heavy-tailed lengths, random turns 5–175°
        n_edges = np.random.randint(3, 9)
        lengths = np.random.pareto(a=1.5, size=n_edges) + 0.2  # heavy-tail
        lengths = (lengths / lengths.sum() * Lk).astype(int)
        # ensure sum = Lk
        if lengths.sum() == 0: lengths[:] = 1
        while lengths.sum() < Lk: lengths[np.random.randint(0, n_edges)] += 1
        while lengths.sum() > Lk:
            i = np.random.randint(0, n_edges)
            if lengths[i] > 1: lengths[i] -= 1
        theta_seg = np.empty(0, dtype=float)
        angle_here = angle_current
        for j, Lj in enumerate(lengths):
            theta_seg = np.concatenate([theta_seg, np.full(Lj, angle_here)])
            if j < n_edges - 1:
                turn = np.deg2rad(np.random.uniform(5, 175)) * (1 if np.random.rand()<0.5 else -1)
                angle_here += turn
        angle_current = theta_seg[-1]

    elif ptype == 'DubinsRS':
        # random word of length 2–3 from {L,R,S} with constant curvature during arcs
        word_len = np.random.randint(2, 4)
        word = np.random.choice(list('LRS'), size=word_len)
        base_sub = np.full(word_len, max(1, Lk // word_len), dtype=int); base_sub[-1] += (Lk - base_sub.sum())
        theta_seg = np.empty(0, dtype=float)
        ang = angle_current
        R = np.random.uniform(0.6, 2.0)  # "radius" proxy via angular rate
        for ch, Lj in zip(word, base_sub):
            tj = np.arange(Lj) * dt
            if ch == 'S':
                seg = np.full(Lj, ang)  # straight heading
            else:
                omega = (1.0/R) * (1 if ch == 'L' else -1) * np.random.uniform(0.8, 1.4)
                if Lj > 1:
                    seg = ang + omega * (tj - tj[0])
                else:
                    seg = np.array([ang + omega * dt])
                ang = seg[-1]
            theta_seg = np.concatenate([theta_seg, seg])
        angle_current = theta_seg[-1]

    elif ptype == 'BezierSpline':
        # single cubic Bézier in local frame, random control points
        t = np.linspace(0, 1, Lk, endpoint=True)
        p0 = np.array([0.0, 0.0])
        p1 = np.random.uniform([-1.0, -1.0], [1.0, 1.0])
        p2 = np.random.uniform([-1.0, -1.0], [1.0, 1.0])
        p3 = np.random.uniform([-1.0, -1.0], [1.0, 1.0])
        Bx = (1-t)**3*p0[0] + 3*(1-t)**2*t*p1[0] + 3*(1-t)*t**2*p2[0] + t**3*p3[0]
        By = (1-t)**3*p0[1] + 3*(1-t)**2*t*p1[1] + 3*(1-t)*t**2*p2[1] + t**3*p3[1]
        dBx = np.gradient(Bx, dt); dBy = np.gradient(By, dt)
        rot = angle_current + np.random.uniform(-np.pi, np.pi)
        theta_seg = (np.arctan2(dBy, dBx) + rot)
        angle_current = theta_seg[-1]

    elif ptype == 'FigureEight':
        # lemniscate (Gerono): x=sin t, y=sin t cos t ; random cycles
        cyc = np.random.uniform(0.8, 2.0)
        t = np.linspace(0, 2*np.pi*cyc, Lk, endpoint=True)
        xg = np.sin(t); yg = np.sin(t)*np.cos(t)
        dxg = np.gradient(xg, dt); dyg = np.gradient(yg, dt)
        rot = angle_current + np.random.uniform(-np.pi, np.pi)
        theta_seg = np.arctan2(dyg, dxg) + rot
        angle_current = theta_seg[-1]

    elif ptype == 'RoseCurve':
        # rhodonea: r = a cos(k t + φ)
        kpetal = np.random.randint(2, 7)
        phi0   = np.random.uniform(-np.pi, np.pi)
        t = np.linspace(0, 2*np.pi, Lk, endpoint=True)
        r = np.cos(kpetal*t + phi0)
        xr = r*np.cos(t); yr = r*np.sin(t)
        dxr = np.gradient(xr, dt); dyr = np.gradient(yr, dt)
        rot = angle_current + np.random.uniform(-np.pi, np.pi)
        theta_seg = np.arctan2(dyr, dxr) + rot
        angle_current = theta_seg[-1]

    elif ptype == 'CycloidFamily':
        # (epi/hypo)trochoid blend; choose at random
        t = np.linspace(0, 2*np.pi*np.random.uniform(1.0, 2.5), Lk, endpoint=True)
        R = np.random.uniform(0.8, 1.6); r = np.random.uniform(0.2, 1.0)
        d  = np.random.uniform(0.1, 1.5)
        if np.random.rand() < 0.5:
            # epicycloid-like
            xcf = (R+r)*np.cos(t) - d*np.cos(((R+r)/r)*t)
            ycf = (R+r)*np.sin(t) - d*np.sin(((R+r)/r)*t)
        else:
            # hypotrochoid-like
            xcf = (R-r)*np.cos(t) + d*np.cos(((R-r)/r)*t)
            ycf = (R-r)*np.sin(t) - d*np.sin(((R-r)/r)*t)
        dx = np.gradient(xcf, dt); dy = np.gradient(ycf, dt)
        rot = angle_current + np.random.uniform(-np.pi, np.pi)
        theta_seg = np.arctan2(dy, dx) + rot
        angle_current = theta_seg[-1]

    elif ptype == 'OUCurvature':
        # OU on angular rate; then integrate to theta
        lam = np.random.uniform(1.0, 4.0)  # mean reversion
        sig = np.random.uniform(2.0, 6.0)  # volatility
        omega = np.zeros(Lk)
        for i in range(1, Lk):
            omega[i] = omega[i-1] + (-lam*omega[i-1])*dt + sig*np.sqrt(dt)*np.random.randn()
        theta_seg = angle_current + np.cumsum(omega)*dt
        angle_current = theta_seg[-1]

    elif ptype == 'CorridorWeave':
        # slow drift + strong oscillation, with grow/decay envelope
        drift_rate = np.deg2rad(np.random.uniform(-25, 25)) / T
        amp0 = np.deg2rad(np.random.uniform(40, 100))
        amp1 = amp0 * np.random.uniform(0.5, 1.2)
        env = np.linspace(amp0, amp1, Lk)
        freq = np.random.uniform(0.3, 1.0)
        phase = np.random.uniform(-np.pi, np.pi)
        theta_seg = angle_current + drift_rate*tloc + env*np.sin(2*np.pi*freq*tloc + phase)
        angle_current = theta_seg[-1]

    elif ptype == 'PursuitEvasion':
        # target follows small Lissajous; heading = angle to target (no full dynamics)
        t = np.linspace(0, 2*np.pi*np.random.uniform(0.8, 1.8), Lk, endpoint=True)
        ax = np.random.uniform(0.5, 1.5); ay = np.random.uniform(0.5, 1.5)
        kx = np.random.randint(1, 4); ky = np.random.randint(1, 4)
        px = np.random.uniform(0, 2*np.pi); py = np.random.uniform(0, 2*np.pi)
        xt = ax*np.sin(kx*t + px)
        yt = ay*np.sin(ky*t + py)
        theta_seg = angle_current + np.arctan2(yt, xt)  # look toward moving target from local origin
        # jitter to avoid locking one side
        theta_seg += np.deg2rad(np.random.uniform(-20, 20))*np.sin(3*t)
        angle_current = theta_seg[-1]

    else:
        theta_seg = np.full(Lk, angle_current)

    # write into theta
    theta[seg_starts[k]:seg_ends[k]] = theta_seg
    primitive_names.append(ptype)

# Anti-bias transforms: random rotation + mirrors
theta += np.random.uniform(0, 2*np.pi)
if np.random.rand() < 0.5:  # mirror y
    theta = np.pi - theta
if np.random.rand() < 0.5:  # mirror x
    theta = -theta

# Slight bias toward stronger net drift (crossing polar axes more)
if np.random.rand() < 0.65:
    th_u = np.unwrap(theta)
    net_turn = th_u[-1] - th_u[0]
    target_abs = np.random.uniform(0.75*np.pi, 2.5*np.pi)  # 135°..450°
    if abs(net_turn) < target_abs:
        extra = (target_abs - abs(net_turn)) * (1.0 if np.random.rand() < 0.5 else -1.0)
        theta += np.linspace(0.0, extra, N)

# Diversity guard: if path barely turns, inject a corrective oscillation
turn_var = np.var(np.gradient(np.unwrap(theta)))
if turn_var < 0.01:
    theta += np.deg2rad(np.random.uniform(30, 90)) * np.sin(2*np.pi*np.random.uniform(0.3, 0.9)*tsim + np.random.uniform(-np.pi, np.pi))

theta = (theta + np.pi) % (2*np.pi) - np.pi  # wrap

# ----------------------------
# 4) Integrate position
# ----------------------------
x = np.cumsum(g * np.cos(theta) * dt)
y = np.cumsum(g * np.sin(theta) * dt)

# ----------------------------
# 5) Tangent heading (for reference)
# ----------------------------
dx = np.gradient(x, dt)
dy = np.gradient(y, dt)
psi_global = np.arctan2(dy, dx)

# ----------------------------
# 6) Heading behavior psi(t)
# ----------------------------
heading_catalog = np.array([
    'tangent', 'fixed', 'spinning', 'oscillating', 'crazy_spin',
    'crab_offset', 'poi_lock', 'scan_hold', 'quantized', 'spin_bursts'
])
heading_weights = np.array([
    1.0, 1.0, 0.9, 0.9, 0.6,
    1.1, 1.0, 0.9, 0.9, 0.8
])
if path_mode == 'solo':
    # small nudge toward fixed/tangent in solo for stability
    heading_weights = heading_weights * np.array([1.2, 1.2, 1.0, 1.0, 0.7, 1.1, 1.0, 0.9, 0.9, 0.8])
heading_probs = heading_weights / heading_weights.sum()
heading_behavior = np.random.choice(heading_catalog, p=heading_probs)

if heading_behavior == 'tangent':
    psi = psi_global

elif heading_behavior == 'fixed':
    fixed_angle_deg = np.round(np.random.uniform(0, 360), 2)
    psi = np.deg2rad(fixed_angle_deg) * np.ones_like(tsim)

elif heading_behavior == 'spinning':
    direction = np.random.choice(['cw', 'ccw'])
    spin_rate_deg = np.random.uniform(30, 360)  # deg/s
    spin_rate = np.deg2rad(spin_rate_deg)
    psi = (-spin_rate if direction == 'cw' else spin_rate) * tsim

elif heading_behavior == 'oscillating':
    angle1_deg = np.random.uniform(0, 360)
    angle2_deg = np.random.uniform(0, 360)
    frequency_hz = np.random.uniform(0.2, 2.0)  # Hz
    amp_rad = np.deg2rad((angle2_deg - angle1_deg) / 2)
    mid_rad = np.deg2rad((angle1_deg + angle2_deg) / 2)
    psi = mid_rad + amp_rad * np.sin(2 * np.pi * frequency_hz * tsim)

elif heading_behavior == 'crazy_spin':
    rotations = np.random.uniform(5.0, 12.0)     # super fast: 5–12 rev over T
    omega = 2*np.pi * rotations / T              # rad/s
    direction = 1.0 if np.random.rand() < 0.5 else -1.0
    phase0 = np.random.uniform(-np.pi, np.pi)
    psi = phase0 + direction * omega * tsim

elif heading_behavior == 'crab_offset':
    # yaw ≈ course + constant bias β, with slow jitter (OU)
    beta = np.deg2rad(np.random.uniform(-25, 25))
    lam  = np.random.uniform(0.3, 1.0); sig = np.deg2rad(np.random.uniform(1.0, 6.0))
    jit = np.zeros_like(tsim)
    for i in range(1, N):
        jit[i] = jit[i-1] + (-lam*jit[i-1])*dt + sig*np.sqrt(dt)*np.random.randn()
    psi = psi_global + beta + jit

elif heading_behavior == 'poi_lock':
    # look at a fixed point (random)
    x0 = np.random.uniform(-5, 5)
    y0 = np.random.uniform(-5, 5)
    psi = np.arctan2(y0 - y, x0 - x)

elif heading_behavior == 'scan_hold':
    # piecewise constant yaw with occasional rapid sweeps
    psi = np.zeros_like(tsim)
    current = np.deg2rad(np.random.uniform(0, 360))
    i = 0
    while i < N:
        # hold
        hold_len = np.random.randint(max(2, int(0.15/dt)), max(3, int(0.45/dt)))
        end = min(N, i+hold_len)
        psi[i:end] = current
        i = end
        if i >= N: break
        # sweep
        sweep_len = np.random.randint(max(2, int(0.05/dt)), max(3, int(0.15/dt)))
        end = min(N, i+sweep_len)
        target = np.deg2rad(np.random.uniform(0, 360))
        if end > i:
            psi[i:end] = np.linspace(current, target, end - i, endpoint=False)
            current = target
        i = end

elif heading_behavior == 'quantized':
    # snap yaw to nearest grid (15° or 30°) with slight bias offset
    base = psi_global + np.deg2rad(np.random.uniform(-20, 20))
    step = np.deg2rad(np.random.choice([15, 30]))
    psi = np.round(base / step) * step

elif heading_behavior == 'spin_bursts':
    # start from path tangent, add 1–3 bursts of fast spin
    psi = psi_global.copy()
    n_b = np.random.randint(1, 4)
    for _ in range(n_b):
        b_len = np.random.randint(max(2, int(0.15/dt)), max(3, int(0.5/dt)))
        b_start = np.random.randint(0, max(1, N - b_len))
        b_end = b_start + b_len
        omega = np.deg2rad(np.random.uniform(180, 720)) * (1 if np.random.rand()<0.5 else -1)
        tloc = (np.arange(b_len) * dt)
        psi[b_start:b_end] += omega * tloc

# wrap heading
psi = (psi + np.pi) % (2*np.pi) - np.pi

# ----------------------------
# 7) Velocities in body frame
# ----------------------------
v_x_global = g * np.cos(psi_global)
v_y_global = g * np.sin(psi_global)
v_x =  v_x_global * np.cos(psi) + v_y_global * np.sin(psi)
v_y = -v_x_global * np.sin(psi) + v_y_global * np.cos(psi)

# ----------------------------
# 8) Quick summary
# ----------------------------
pmode_desc = "SOLO" if path_mode == 'solo' else "Hybrid:" + "→".join(primitive_names)
print(
    "Wind: {:.2f} m/s @ {:.2f}° | Speed: {} (≈[{:.2f},{:.2f}] m/s) | "
    "Heading: {} | Path: {}".format(
        w_mag, zeta_deg, speed_behavior, g.min(), g.max(),
        heading_behavior, pmode_desc
    )
)
```
